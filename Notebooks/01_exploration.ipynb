```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load only first 100,000 rows to start
df = pd.read_csv('E:/federal-contract-dashboard/Data/FY2025_097_Contracts_Full_20260610_1.csv',
                 nrows=100000,
                 low_memory=False)

# First look
print("Shape:", df.shape)
print("\nNumber of columns:", len(df.columns))
print("\nFirst 5 rows:")
print(df.head())
```

```python
# See all the column names to pick the important ones
for col in df.columns:
    print(col)
```

```python
# Keep only the columns we need for our story
columns_to_keep = [
    'action_date',
    'action_date_fiscal_year',
    'federal_action_obligation',
    'current_total_value_of_award',
    'awarding_sub_agency_name',
    'recipient_name',
    'recipient_state_name',
    'naics_description',
    'product_or_service_code_description',
    'extent_competed',
    'type_of_contract_pricing',
    'women_owned_small_business',
    'veteran_owned_business',
    'service_disabled_veteran_owned_business',
    'small_disadvantaged_business',
    'minority_owned_business'
]

df_clean = df[columns_to_keep].copy()

print("Shape:", df_clean.shape)
df_clean.head()
```

```python
# Check for missing values
print(df_clean.isnull().sum())

# Check data types
print("\n")
print(df_clean.dtypes)
```

```python
# Convert action_date to a proper date
df_clean['action_date'] = pd.to_datetime(df_clean['action_date'])

# Confirm it worked
print(df_clean['action_date'].dtype)
print(df_clean['action_date'].min(), "to", df_clean['action_date'].max())
```

```python
import pandas as pd

chunk_size = 50000  # read 50,000 rows at a time
chunks = []

for chunk in pd.read_csv('E:/federal-contract-dashboard/Data/FY2025_097_Contracts_Full_20260610_1.csv',
                          usecols=columns_to_keep,
                          chunksize=chunk_size,
                          low_memory=False):
    chunks.append(chunk)

df_full = pd.concat(chunks, ignore_index=True)

print("Shape:", df_full.shape)
```

```python
df_full['action_date'] = pd.to_datetime(df_full['action_date'])
print("Date range:", df_full['action_date'].min(), "to", df_full['action_date'].max())
print("\nFiscal years included:")
print(df_full['action_date_fiscal_year'].value_counts().sort_index())
```

```python
# Drop rows with missing state (small % loss)
df_full = df_full.dropna(subset=['recipient_state_name'])

# Fill missing extent_competed
df_full['extent_competed'] = df_full['extent_competed'].fillna('Not Specified')

print("Final shape:", df_full.shape)
print("\nRemaining missing values:")
print(df_full.isnull().sum())
```

```python
# Total obligated dollars
total_obligated = df_full['federal_action_obligation'].sum()
print(f"Total federal obligation: ${total_obligated:,.2f}")

# Breakdown by business category
categories = ['veteran_owned_business', 'service_disabled_veteran_owned_business',
              'women_owned_small_business', 'minority_owned_business',
              'small_disadvantaged_business']

for cat in categories:
    amount = df_full[df_full[cat] == 'Y']['federal_action_obligation'].sum()
    pct = (amount / total_obligated) * 100
    print(f"{cat}: ${amount:,.2f} ({pct:.2f}%)")
```

```python
categories = ['veteran_owned_business', 'service_disabled_veteran_owned_business',
              'women_owned_small_business', 'minority_owned_business',
              'small_disadvantaged_business']

for cat in categories:
    print(f"\n{cat}:")
    print(df_full[cat].value_counts())
```

```python
total_obligated = df_full['federal_action_obligation'].sum()
print(f"Total federal obligation: ${total_obligated:,.2f}\n")

categories = ['veteran_owned_business', 'service_disabled_veteran_owned_business',
              'women_owned_small_business', 'minority_owned_business',
              'small_disadvantaged_business']

for cat in categories:
    amount = df_full[df_full[cat] == 't']['federal_action_obligation'].sum()
    pct = (amount / total_obligated) * 100
    print(f"{cat}: ${amount:,.2f} ({pct:.2f}%)")
```

```python
# Top 10 vendors by total obligation
top_vendors = df_full.groupby('recipient_name')['federal_action_obligation'].sum().sort_values(ascending=False).head(10)
print(top_vendors)
```

```python
# Standardize company name variations
df_full['recipient_name_clean'] = df_full['recipient_name'].str.upper().str.strip()
df_full['recipient_name_clean'] = df_full['recipient_name_clean'].str.replace(r'\bCORP\b', 'CORPORATION', regex=True)
df_full['recipient_name_clean'] = df_full['recipient_name_clean'].str.replace(r'\bINC\b\.?', 'INCORPORATED', regex=True)

# Re-run top vendors with cleaned names
top_vendors_clean = df_full.groupby('recipient_name_clean')['federal_action_obligation'].sum().sort_values(ascending=False).head(10)
print(top_vendors_clean)
```

```python
import pandas as pd

# All 5 FY2025 file paths
file_paths = [
    'E:/federal-contract-dashboard/Data/FY2025_097_Contracts_Full_20260610_1.csv',
    'E:/federal-contract-dashboard/Data/FY2025_097_Contracts_Full_20260610_2.csv',
    'E:/federal-contract-dashboard/Data/FY2025_097_Contracts_Full_20260610_3.csv',
    'E:/federal-contract-dashboard/Data/FY2025_097_Contracts_Full_20260610_4.csv',
    'E:/federal-contract-dashboard/Data/FY2025_097_Contracts_Full_20260610_5.csv',
]

columns_to_keep = [
    'action_date',
    'action_date_fiscal_year',
    'federal_action_obligation',
    'current_total_value_of_award',
    'awarding_sub_agency_name',
    'recipient_name',
    'recipient_state_name',
    'naics_description',
    'product_or_service_code_description',
    'extent_competed',
    'type_of_contract_pricing',
    'women_owned_small_business',
    'veteran_owned_business',
    'service_disabled_veteran_owned_business',
    'small_disadvantaged_business',
    'minority_owned_business'
]

chunk_size = 50000
all_chunks = []

for path in file_paths:
    print(f"Loading {path} ...")
    for chunk in pd.read_csv(path, usecols=columns_to_keep, chunksize=chunk_size, low_memory=False):
        # Keep only meaningful transactions
        chunk = chunk[chunk['federal_action_obligation'] != 0]
        all_chunks.append(chunk)
    print(f"Done with {path}")

df_all = pd.concat(all_chunks, ignore_index=True)
print("\nFinal combined shape:", df_all.shape)
```

```python
df_all['action_date'] = pd.to_datetime(df_all['action_date'])
print("Date range:", df_all['action_date'].min(), "to", df_all['action_date'].max())
print("\nFiscal years included:")
print(df_all['action_date_fiscal_year'].value_counts().sort_index())
```

```python
# Clean up missing data
df_all = df_all.dropna(subset=['recipient_state_name'])
df_all['extent_competed'] = df_all['extent_competed'].fillna('Not Specified')
print("Final shape:", df_all.shape)
```

```python
# How much money went to each business category
total_obligated = df_all['federal_action_obligation'].sum()
print(f"Total spent: ${total_obligated:,.2f}\n")

categories = ['veteran_owned_business', 'service_disabled_veteran_owned_business',
              'women_owned_small_business', 'minority_owned_business',
              'small_disadvantaged_business']

for cat in categories:
    amount = df_all[df_all[cat] == 't']['federal_action_obligation'].sum()
    pct = (amount / total_obligated) * 100
    print(f"{cat}: ${amount:,.2f} ({pct:.2f}%)")
```

```python
# Top 10 companies that got the most money
df_all['recipient_name_clean'] = df_all['recipient_name'].str.upper().str.strip()
df_all['recipient_name_clean'] = df_all['recipient_name_clean'].str.replace(r'\bCORP\b', 'CORPORATION', regex=True)
df_all['recipient_name_clean'] = df_all['recipient_name_clean'].str.replace(r'\bINC\b\.?', 'INCORPORATED', regex=True)

top_vendors = df_all.groupby('recipient_name_clean')['federal_action_obligation'].sum().sort_values(ascending=False).head(10)
print(top_vendors)
```

```python
# Saving the clean dataset to csv file 
df_all.to_csv('E:/federal-contract-dashboard/Data/cleaned_fy2025_full.csv', index=False)
print("Saved!")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Reload the data fresh with all 3 columns needed
df = pd.read_csv('E:/federal-contract-dashboard/Data/cleaned_fy2025_full.csv',
                 usecols=['federal_action_obligation',   # dollar amount
                          'awarding_sub_agency_name',    # which DoD branch
                          'recipient_name_clean'],       # which company
                 low_memory=False)

# Convert text columns to numbers (ML needs numbers not text)
le_agency = LabelEncoder()
le_vendor = LabelEncoder()

df['agency_encoded'] = le_agency.fit_transform(df['awarding_sub_agency_name'])
df['vendor_encoded'] = le_vendor.fit_transform(df['recipient_name_clean'])

# Select features for the model
features = df[['federal_action_obligation',  # dollar amount
               'agency_encoded',              # agency as number
               'vendor_encoded']]             # vendor as number

print("Features ready:", features.shape)
print(features.head(3))
```
